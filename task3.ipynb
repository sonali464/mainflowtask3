```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("C:\\Users\\subha\\OneDrive\\Desktop\\mainflow\\householdtask3.csv")
data
```

```python
display(data.head(10))
```

```python
#scatter plot with year against own
plt.scatter(data['year'],data['own'])
#adding title to plot
plt.title("Scatter Plot")
#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')
#show result
plt.show()
```

```python
#Line chart with year against own
plt.plot(data['year'])
plt.plot(data['own'])
plt.title("Line Chart")
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
#bar plot
plt.bar(data['year'],data['income'])
plt.title("Bar chart")
plt.xlabel('year')
plt.ylabel('income')
plt.show()
```

```python
#histogram
plt.hist(data['expenditure'])
plt.xlabel("expenditure")
plt.title("Histogram")
plt.show()
```
